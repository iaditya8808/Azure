```python
# Scenario
# A student is building a simple AI assistant in Google Colab for a class project. The assistant is
# hosted on Azure Foundry, and the student wants to ask it questions like:

# “What can you help with?”

# “Explain this topic simply.”

# “Give me a short summary.”

# Instead of talking directly to the model through a chat app, the student uses Python code to connect to the Azure service.

from google.colab import userdata

api_key = userdata.get("AZURE_OPENAI_API_KEY")
print("Loaded key:", api_key[:6] + "..." if api_key else "Missing")

from google.colab import userdata
from openai import OpenAI

my_endpoint = "https://student-ai-001.services.ai.azure.com/api/projects/proj-default"
my_agent = "email-assistant"
my_version = "2"

api_key = userdata.get("AZURE_OPENAI_API_KEY")

openai_client = OpenAI(
    base_url=f"{my_endpoint}/openai/v1",
    api_key=api_key
)

response = openai_client.responses.create(
    input=[{"role": "user", "content": "Tell me what you can help with."}],
    extra_body={
        "agent_reference": {
            "name": my_agent,
            "version": my_version,
            "type": "agent_reference"
        }
    },
)

print(response.output_text)
```

```python
from google.colab import userdata
from openai import OpenAI

api_key = userdata.get("API_KEY")

my_endpoint = "https://customer-chatbot1.services.ai.azure.com/api/projects/proj-default"
my_agent = "AI-CustomerChatBot"
my_version = "2"

client = OpenAI(
    base_url=f"{my_endpoint}/openai/v1",
    api_key=api_key
)

user_query = input("Ask your question: ")

# API call
response = client.responses.create(
    input=[
        {
            "role": "user",''
            "content": user_query
        }
    ],

    extra_body={
        "agent_reference": {
            "name":my_agent,
            "version": my_version,
            "type": "agent_reference"
        }
    },
)

print("\nBot:", response.output_text)
```

```python
# AI Interview Simulator with Feedback

# Concept: Simulates real interviews

# 💡 Output:

# Questions
# Feedback on answers
# Improvement tips

# Twist:
# Add HR + Technical + Stress interview modes
```

```python
from google.colab import userdata
from openai import OpenAI

# Load API key
api_key = userdata.get("API_KEY")

# Azure Agent details
my_endpoint = "https://customer-chatbot1.services.ai.azure.com/api/projects/proj-default"
my_agent = "AI-Interview-Simulator"
my_version = "2"

# Initialize client
client = OpenAI(
    base_url=f"{my_endpoint}/openai/v1",
    api_key=api_key
)

#  Select Interview Mode
print("Select Interview Mode:")
print("1. HR")
print("2. Technical")
print("3. Stress")

mode_input = input("Enter choice (1/2/3): ")

if mode_input == "1":
    mode = "HR Interview"
elif mode_input == "2":
    mode = "Technical Interview"
else:
    mode = "Stress Interview"

#  Ask user to answer
user_answer = input(f"\n[{mode}] Answer the question: Tell me about yourself\n\nYour Answer: ")

#  Prompt (passed as part of input since agent blocks instructions)
full_prompt = f"""
You are an AI Interview Simulator.

Mode: {mode}

Task:
1. Evaluate the candidate's answer
2. Give:
   - Score (1-10)
   - Strengths
   - Weaknesses
   - Improvement Tips

Candidate Answer:
{user_answer}
"""

#  API Call
response = client.responses.create(
    input=full_prompt,
    extra_body={
        "agent_reference": {
            "name": my_agent,
            "version": my_version,
            "type": "agent_reference"
        }
    }
)

#  Output
print("\n Feedback:\n")
print(response.output_text)
```
