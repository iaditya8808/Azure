```python
from google.colab import userdata
from openai import OpenAI

# 🔐 Load API Key securely from Colab Secrets
api_key = userdata.get("OPENAI_API_KEY")

# ✅ Azure resource endpoint and deployment
endpoint = "https://chat-bot-01-resource.openai.azure.com"
deployment_name = "gpt-4.1"

# Initialize client with proper Azure endpoint and API version
client = OpenAI(
    api_key=api_key,
    base_url=f"{endpoint}/openai/deployments/{deployment_name}/",
    default_query={"api-version": "2024-02-01"}  # required for Azure
)

# Create a chat completion
completion = client.chat.completions.create(
    model=deployment_name,
    messages=[
        {
            "role": "user",
            "content": "What is the capital of France?",
        }
    ],
)

# ✅ Access the message content correctly
print(completion.choices[0].message.content)
```

```python
import gradio as gr

def ask_ai_healthcare(user_query):
    """
    Sends a user query about AI in healthcare to the Azure OpenAI model
    and returns the model's response.
    """
    try:
        # Ensure 'client' and 'deployment_name' are accessible from previous cells
        global client, deployment_name

        response = client.chat.completions.create(
            model=deployment_name,
            messages=[
                {"role": "system", "content": "You are an AI assistant specialized in healthcare. Provide accurate and concise information about AI's applications, challenges, and ethical considerations in the healthcare domain."},
                {"role": "user", "content": user_query},
            ],
            temperature=0.7,
            max_tokens=500
        )
        return response.choices[0].message.content
    except NameError:
        return "Error: 'client' or 'deployment_name' not defined. Please run the setup cells first."
    except Exception as e:
        return f"An error occurred: {e}"

# Create the Gradio interface
iface = gr.Interface(
    fn=ask_ai_healthcare,
    inputs=gr.Textbox(lines=5, label="Ask about AI in Healthcare"),
    outputs="text",
    title="AI in Healthcare Q&A with Azure OpenAI",
    description="Ask any question about the applications, challenges, or ethics of AI in the healthcare sector."
)

# Launch the Gradio interface
iface.launch(debug=True)
```
